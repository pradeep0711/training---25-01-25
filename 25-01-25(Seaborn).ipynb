```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
sns.barplot(data = df,x='Pclass',y='Survived',errorbar = None)
plt.title("Survival Rate by Passenger Class")
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data = df,x='Age',y='Fare',hue='Survived',palette='magma')
plt.title("Age vs Fare")
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
p_c = df[['Age', 'Fare', 'Pclass','Survived']] 
sns.pairplot(data=p_c, hue='Survived',palette = 'deep',diag_kind = 'kde') 
plt.suptitle('Pairwise Relationships',y = 1.1)
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="magma", fmt=".2f",linewidth= 0.5)
plt.title("Correlation HeatMap")
plt.show()
```
