```python
df['Avg_Temp'] = df['AverageTemperature'].mean()
df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Increase the path chunk size for rendering large datasets
plt.rcParams['agg.path.chunksize'] = 10000  # Adjust this as needed

# Convert 'dt' to datetime format
df['dt'] = pd.to_datetime(df['dt'], errors='coerce')

# Extract the year from the datetime column
df['Year_only'] = df['dt'].dt.year

# Plot the graph
plt.plot(df['Year_only'], df['AverageTemperature'])
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.title('Average Temperature Over Time')
plt.show()
```

```python
df['AverageTemperature'] = df['AverageTemperature'].fillna(df['AverageTemperature'].mean())
df
```

```python
df.info()
```

```python
city_data = df[df['City']=='New York'].copy()
city_data['dt'] = pd.to_datetime(city_data['dt'],errors='coerce')
city_data['Year'] = city_data['dt'].dt.year
ann_avg_temp = city_data.groupby('Year')['AverageTemperature'].mean().reset_index()
plt.figure(figsize=(12,6))
plt.plot(ann_avg_temp['Year'],ann_avg_temp['AverageTemperature'],linestyle='-',color='blue',marker = 'P')
plt.title('Average Trend In New York ')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
```

```python
country_avg_temp = df.groupby('Country')['AverageTemperature'].mean().reset_index()
country_avg_temp = country_avg_temp.dropna()
t_c = country_avg_temp.sort_values(by ='AverageTemperature',ascending = False).head(10)
c = ['lightblue','lightpink','lightgreen','orange']
plt.figure(figsize=(12,6))
plt.bar(t_c['Country'],t_c['AverageTemperature'],color = c )
plt.title('Top 10 Countries with the highest Average')
plt.xlabel('country')
plt.ylabel('Avg Temp')
plt.xticks(rotation = 45)
plt.show()
```

```python
c = 60
plt.figure(figsize=(10,6))
plt.hist(df['AverageTemperature'],bins = c,edgecolor='black',color = 'lightblue')
plt.title('Average Temperature Across all data')
plt.xlabel('Temperature')
plt.ylabel('Frquency')
plt.grid(axis='y',linestyle ='--',alpha = 0.5)
plt.show()
```

```python
avg_temp = df['AverageTemperature'].head(2000)
temp_uncertainty = df['AverageTemperatureUncertainty'].head(2000)
plt.figure(figsize=(12,8))
plt.scatter(avg_temp,temp_uncertainty,color="green",s= 10, edgecolor="black",alpha = 0.5)
#plt.xticks(range(1,6,1))
#plt.yticks(range(10,60,10))
plt.title("Scatter Plot")
plt.xlabel("Avg Temp")
plt.ylabel("Temp Uncertainty")
plt.show()
```

```python
country_data = df[df['Country'] == 'India'].copy()
country_data['dt'] = pd.to_datetime(country_data['dt'], errors='coerce')
country_data = country_data.dropna(subset=['dt'])
country_data['Year'] = country_data['dt'].dt.year
country_data = country_data[['Year', 'AverageTemperature']].dropna()
grouped_data = [group['AverageTemperature'] for _, group in country_data.groupby('Year')]
plt.boxplot(grouped_data,showfliers = True,patch_artist = True)
plt.xlabel('Year Or Country')
plt.ylabel('Average Temperature')
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data = tips,x='total_bill',kde=True,bins=50,color= 'red')
plt.title('Distribution of tables')
plt.show()
```

```python
sns.kdeplot(data = tips,x='total_bill',fill = True ,color= 'red',log_scale = True)
```

```python
sns.barplot(data = tips,x="day",y="total_bill",palette="viridis",hue='sex')
plt.title('Bar Plot ')
```

```python
sns.boxplot(data = tips,x="day",y="total_bill",palette="plasma",hue='sex')
plt.title('Box PLot of Total Bills by Day And Gender')
```

```python
sns.scatterplot(data = tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("ScatterPlot of total bills vs Tip")
```

```python
#Line Plot
sns.lineplot(data = tips,x="size",y="tip",marker="o",style="time",color ='green')
plt
```

```python
sns.pairplot(data = tips,hue="sex",palette = "husl")
plt.show()
```

```python
# Select only numeric columns
numeric_columns = tips.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="magma", fmt=".2f")
plt.title("Correlation HeatMap")
plt.show()
```
